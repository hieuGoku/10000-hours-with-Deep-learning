## Close form

```python
def find_theta (X, y):
    return (np.linalg.inv(X.T @ X) @ X.T) @ y
```

```python
X = np.concatenate((x, np.ones((len(x), 1))), axis=1)
X.shape, y.shape
```

```python
w = find_theta(X, y)
w
```

```python
plt.plot(X@w, label= 'predict')
plt.plot(y, label= 'target')
plt.show()
```

```python
np.mean((X@w - y)**2)
```

## Gradient 

```python
x.shape, y.shape
```

```python
n_sample, n_feature = x.shape
n_sample, n_feature
```

```python
X = np.concatenate((x, np.ones((len(x), 1))), axis=1)
```

```python
X.shape, y.shape
```

```python
w = np.random.rand(n_feature+1, 1)
n_epoch = 500
lr = 0.0000001
losses = []
bs = n_sample
```

```python
w.shape
```

```python
for _ in range(n_epoch):
    for i in range(0, n_sample, bs):

        xi = X[i:i+bs, :]
        yi = y[i:i+bs]

        y_pred = xi @ w

        loss = (y_pred - yi)**2
        losses.append(loss.mean())

        dw = xi.T @ (y_pred - yi) * 2/bs

        w = w - lr*dw
```

```python
plt.plot(losses)
plt.show()
```
