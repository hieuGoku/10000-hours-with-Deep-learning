```python
import numpy as np
import tensorflow as tf
import tensorflow.keras as keras

data = np.random.randint(0, 256, (28, 28, 1))
print(data.shape)
print(type(data))
```

```python
data_tf = tf.convert_to_tensor(data, dtype=tf.float32)
print(data_tf.shape)
print(type(data_tf))
```

```python
conv_layer = keras.layers.Conv2D(filters=1, 
                                kernel_size=(3, 3))

output = conv_layer(data_tf)
print(output.shape)
```

```python
data_tf = tf.reshape(data_tf, (1, 28, 28, 1))
print(data_tf.shape)
```

```python
conv_layer = keras.layers.Conv2D(filters=1, 
                                kernel_size=(3, 3),
                                activation='relu')
output = conv_layer(data_tf)
print(output.shape)
print(output[0, :,:,0])
```

```python
weights = conv_layer.get_weights()
print(type(weights))
print(len(weights))
```

```python
print(weights[0].shape)
```

```python
print(weights[0].shape)
print(weights[1].shape)
```

```python
print(weights[0][:,:,0,0])
print(weights[1])
```

## stride

```python
data_tf = tf.reshape(data_tf, (1, 28, 28, 1))
print(data_tf.shape)
```

```python
conv_layer = keras.layers.Conv2D(filters=5, 
                                kernel_size=(3, 3), 
                                strides=2)
output = conv_layer(data_tf)
print(output.shape)
```

```python
conv_layer = keras.layers.Conv2D(filters=1, 
                                kernel_size=(4, 4), 
                                strides=2)
output = conv_layer(data_tf)
print(output.shape)
```

```python
conv_layer = keras.layers.Conv2D(filters=1, kernel_size=(5, 5), strides=2)
output = conv_layer(data_tf)
print(output.shape)
```
